```python
from scapy.all import sniff, IP, TCP, UDP, ICMP

# Function to process each captured packet
def process_packet(packet):
    print("\n==============================")

    # Check if packet has IP layer
    if packet.haslayer(IP):
        ip_layer = packet[IP]
        print(f"Source IP      : {ip_layer.src}")
        print(f"Destination IP : {ip_layer.dst}")
        print(f"Protocol       : {ip_layer.proto}")

        # Identify protocol type
        if packet.haslayer(TCP):
            print("Protocol Name  : TCP")
        elif packet.haslayer(UDP):
            print("Protocol Name  : UDP")
        elif packet.haslayer(ICMP):
            print("Protocol Name  : ICMP")
        else:
            print("Protocol Name  : Other")

        # Print payload if available
        if packet.haslayer(TCP) or packet.haslayer(UDP):
            payload = bytes(packet.payload)
            if payload:
                print("Payload        :", payload[:50])  # Print first 50 bytes

# Start sniffing
print("Starting packet capture... Press Ctrl+C to stop.")
sniff(prn=process_packet, count=10)  # Capture 10 packets
```
