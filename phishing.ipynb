```python
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.enum.text import PP_ALIGN
from pptx.dml.color import RGBColor
from pptx.enum.shapes import MSO_AUTO_SHAPE_TYPE

# Create Presentation
prs = Presentation()

def add_slide(title, content):
    slide_layout = prs.slide_layouts[1]
    slide = prs.slides.add_slide(slide_layout)
    slide.shapes.title.text = title

    content_placeholder = slide.placeholders[1]
    content_placeholder.text = content[0]

    for line in content[1:]:
        p = content_placeholder.text_frame.add_paragraph()
        p.text = line
        p.level = 1

# Slide 1: Title
slide_layout = prs.slide_layouts[0]
slide = prs.slides.add_slide(slide_layout)
slide.shapes.title.text = "Phishing Awareness Training"
slide.placeholders[1].text = "Learn How to Identify and Prevent Phishing Attacks"

# Slide 2: Introduction to Phishing
add_slide("What is Phishing?",
[
"Phishing is a cyber attack where attackers impersonate trusted entities.",
"It aims to steal sensitive data like passwords, OTPs, and banking details.",
"Delivered via emails, SMS, social media, or fake websites."
])

# Slide 3: Types of Phishing
add_slide("Common Types of Phishing",
[
"Email Phishing – Fake emails from banks or companies.",
"Spear Phishing – Targeted attack on specific individuals.",
"Smishing – Phishing via SMS messages.",
"Vishing – Voice call scams."
])

# Slide 4: Recognizing Phishing Emails
add_slide("How to Recognize Phishing Emails",
[
"Suspicious sender address.",
"Urgent or threatening language.",
"Spelling and grammar mistakes.",
"Unexpected attachments or links."
])

# Slide 5: Identifying Fake Websites
add_slide("How to Identify Fake Websites",
[
"Check URL carefully (misspellings like amaz0n.com).",
"Look for HTTPS and padlock symbol.",
"Poor website design or copied content.",
"Requests for confidential information."
])

# Slide 6: Social Engineering Tactics
add_slide("Social Engineering Tactics",
[
"Creating urgency ('Your account will be locked!').",
"Pretending to be authority (Bank, Police, IT Department).",
"Building fake trust.",
"Using fear or curiosity."
])

# Slide 7: Real-World Example
add_slide("Real-World Example",
[
"Fake email from 'Bank Support' asking to verify account.",
"Link redirects to fake login page.",
"User enters credentials.",
"Attacker steals banking details."
])

# Slide 8: Best Practices to Stay Safe
add_slide("Best Practices to Avoid Phishing",
[
"Never click suspicious links.",
"Verify sender before responding.",
"Use strong and unique passwords.",
"Enable Two-Factor Authentication (2FA).",
"Keep antivirus and system updated."
])

# Slide 9: Interactive Quiz 1
add_slide("Quiz Time!",
[
"1. Which is a sign of phishing?",
"A) Professional domain email",
"B) Urgent message asking for password",
"C) Secure website with valid certificate",
"Correct Answer: B"
])

# Slide 10: Interactive Quiz 2
add_slide("Quiz Time!",
[
"2. What should you do if you receive a suspicious email?",
"A) Click the link immediately",
"B) Report and delete it",
"C) Share it with friends",
"Correct Answer: B"
])

# Save File
file_name = "Phishing_Awareness_Training.pptx"
prs.save(file_name)

print("Presentation created successfully!")
```
